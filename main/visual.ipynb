```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot(monte, qcs, gacs, id):
  plt.figure(figsize=(40,5))
  plt.subplot(1, 4, id)
  plt.plot(monte['Time'], Monte['Nodes'], color='red', marker='o')
  plt.plot(qcs['Time'], qcs['Nodes'], color='green', marker='o')
  plt.plot(gacs['Time'], gacs['Nodes'], color='blue', marker='o')

  plt.title('Result', fontsize=14)
  plt.xlabel('Time(s)', fontsize=14)
  plt.ylabel('Dead nodes', fontsize=14)

  plt.legend(['Monte', 'QCS', 'GACS'])
  plt.grid(True)
  # plt.show()
```

```python
def plot2(monte, qcs, gacs, id):
  plt.figure(figsize=(40,5))
  plt.subplot(1, 4, id)
  plt.plot(monte['Time'], Monte['Nodes'], color='red', marker='o')
  plt.plot(qcs['Time'], qcs['Nodes'], color='green', marker='o')
  plt.plot(gacs['Time'], gacs['Nodes'], color='blue', marker='o')

  plt.title('Result', fontsize=14)
  plt.xlabel('Time(s)', fontsize=14)
  plt.ylabel('Sum Energy', fontsize=14)

  plt.legend(['Monte', 'QCS', 'GACS'])
  plt.grid(True)
  # plt.show()
```
